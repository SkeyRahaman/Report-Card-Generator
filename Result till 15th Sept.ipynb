```python
import numpy as np
import pandas as pd
```

```python
users=pd.read_csv('users - users.csv')
marks=pd.read_csv('evaluation - evaluation.csv')
myday=pd.read_csv('myday - myday.csv')
gyan=pd.read_csv('gyan - gyan.csv')
```

## Top 10 Performers

```python
myday.drop_duplicates(subset=['user_id','date'],inplace=True)
myday_frequency=myday['user_id'].value_counts().reset_index()
myday_frequency.rename(columns={'index':'user_id','user_id':'marks'},inplace=True)
myday_final=myday_frequency.merge(users,on='user_id')[['fname','marks']]
myday_final
```

```python
gyan.drop_duplicates(subset=['user_id','date'],inplace=True)
gyan_frequency=gyan['user_id'].value_counts().reset_index()
gyan_frequency.rename(columns={'index':'user_id','user_id':'marks'},inplace=True)
final_gyan=gyan_frequency.merge(users,on='user_id')[['fname','marks']]
final_gyan
```

```python
total_marks=marks.groupby('user_id')['marks'].sum().reset_index()
final_marks=total_marks.merge(users,on='user_id')[['fname','marks']]
```

```python
final=pd.concat([final_marks,final_gyan,myday_final])
```

```python
final.groupby('fname')['marks'].sum().sort_values(ascending=False).head(11)
```

## Most Consistent

```python
mc=pd.concat([myday_final,final_gyan])
mc=mc.groupby('fname')['marks'].sum().reset_index()
late=marks.groupby('user_id')['late'].sum().reset_index()
final_late=late.merge(users,on='user_id')[['fname','late']]
consistent=mc.merge(final_late,on='fname')
consistent['score']=consistent['marks']-consistent['late']
```

```python
consistent.sort_values('score',ascending=False).head(1)
```

## Most Task Wins

```python
x=marks.groupby('user_id')['winner'].sum().reset_index()
x=x.merge(users,on='user_id')[['fname','winner']]
x.sort_values('winner',ascending=False).head(1)
```
